Import Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
```

```python
df= pd.read_csv(r".\Titanic-Dataset.csv")
```

removing 'Cabin' column due to have more number of null values and the data it contains is nominal which will not be treated

```python
df = df.drop(['PassengerId','Cabin','Name','Ticket'],axis=1)
```

```python
df.replace({'Sex':{'male':1,'female':0},
             'Embarked':{'S':0,'C':1,'Q':2}
           },inplace=True)
```

```python
df
```

Treating null values

```python
df.fillna({'Age':df['Age'].mean()},inplace=True) #filling nan value with the mean as it is numerical column
df.fillna({'Embarked':df['Embarked'].mode()[0]},inplace=True) #filling nan value with mode as it is categorical column
```

```python
df.info()
```

preparing data for spliting data in train and test

```python
X= df.drop("Survived",axis=1)
y=df['Survived']
```

splitting the data in to train data and test data

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

training logistic regression model as it is a predictive classification problem

```python
logistic_reg = LogisticRegression(max_iter=500)
```

```python
logistic_reg.fit(X_train,y_train)
```

```python
logistic_reg.score(X_train,y_train)
```

testing the logistic reggression

```python
y_pred = logistic_reg.predict(X_test)
```

Creating the classification report

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
